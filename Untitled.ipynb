## k-nearest Neighbors

```python
from sklearn import neighbors
import pandas as pd
from sklearn.metrics import classification_report
```

```python
data = plant1_nonan
data.head()
```

```python
data.columns
```

```python
# feature : 2~12  
# label : 13 14 15
# day  :  16 17 18 19 20   year ~  minute
```

```python
data.info()
```

```python
data[['plant1_train.cond_loc1', 'plant1_train.cond_loc2',
       'plant1_train.cond_loc3']].astype('category')
```

```python
X_feature = data.iloc[:,2:13]
X_day = data.iloc[:, 16:21]
X = pd.concat([X_feature, X_day], axis=1)
X.head()
```

```python
y = data[['plant1_train.cond_loc1', 'plant1_train.cond_loc2',
       'plant1_train.cond_loc3']]
y.head()
```

```python
print(X.shape, y.shape)
```

```python
kNN = neighbors.KNeighborsClassifier()
kNN.fit(X,y)

Z= kNN.predict(X)
target_names = ['loc1','loc2','loc3']
print(classification_report(y, Z, target_names=target_names))
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, train_size=0.8, shuffle=False)
```

```python
KNN = neighbors.KNeighborsClassifier()
KNN.fit(X_train, y_train)

pred = KNN.predict(X_test)
target_names = ['loc1', 'loc2', 'loc3']
print(classification_report(y_test, pred, target_names = target_names))
```
