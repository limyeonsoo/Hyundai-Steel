```python
#validation???
def data_loader(url1, url2):
    data1 = pd.read_csv(url1,parse_dates=['date'])
    data2 = pd.read_csv(url2,parse_dates=['date'])
    # 쓸데없는 컬럼 삭제 일단 Unnamed
    
    label_24 = ['X24H_cond_loc1','X24H_cond_loc2','X24H_cond_loc3']
    label_48 = ['X48H_cond_loc1','X48H_cond_loc2','X48H_cond_loc3']
    
    data1["month"] = data1["date"].dt.month
    data1["day"] = data1["date"].dt.day
    data1["hour"] = data1["date"].dt.hour
    data2["month"] = data2["date"].dt.month
    data2["day"] = data2["date"].dt.day
    data2["hour"] = data2["date"].dt.hour
    
    feature = ['tem_loc1','hum_loc1','coil_loc1',
               'tem_loc2','hum_loc2','coil_loc2',
               'tem_loc3','hum_loc3','coil_loc3',
               'tem_out','hum_out','month','day','hour']
    X1 = Normalizing.normalize(data1, feature)
    X2 = Normalizing.normalize(data2, feature)
    
    X1 = X1[feature]
    X2 = X2[feature]
    
    y_1024 = data1[label_24]
    y_1024 = y_1024.dropna(axis=0)
    y_2024 = data2[label_24]
    y_2024 = y_2024.dropna(axis=0)
    
    X_1024 = X1.iloc[:y_1024.shape[0], :]
    X_2024 = X2.iloc[:y_2024.shape[0], :]
    
    y_1048 = data1[label_48]
    y_1048 = y_1048.dropna(axis=0)
    y_2048 = data2[label_48]
    y_2048 = y_2048.dropna(axis=0)
    
    X_1048 = X1.iloc[:y_1048.shape[0], :]
    X_2048 = X2.iloc[:y_2048.shape[0], :]
    # 행의 개수가 다르기 때문.
    X_1124 = X_1024[['tem_loc1','hum_loc1','coil_loc1','tem_out','hum_out','month','day','hour']]
    X_1124.columns=['tem_in','hum_in','tem_coil','tem_out_loc1','hum_out_loc1','month','day','hour']
    y_1124 = y_1024['X24H_cond_loc1']
    
    X_1148 = X_1048[['tem_loc1','hum_loc1','coil_loc1','tem_out','hum_out','month','day','hour']]
    X_1148.columns=['tem_in','hum_in','tem_coil','tem_out_loc1','hum_out_loc1','month','day','hour']
    y_1148 = y_1048['X48H_cond_loc1']
    
    X_1224 = X_1024[['tem_loc2','hum_loc2','coil_loc2','tem_out','hum_out','month','day','hour']]
    X_1224.columns=['tem_in','hum_in','tem_coil','tem_out_loc1','hum_out_loc1','month','day','hour']
    y_1224 = y_1024['X24H_cond_loc2']
        
    X_1248 = X_1048[['tem_loc2','hum_loc2','coil_loc2','tem_out','hum_out','month','day','hour']]
    X_1248.columns=['tem_in','hum_in','tem_coil','tem_out_loc1','hum_out_loc1','month','day','hour']
    y_1248 = y_1048['X48H_cond_loc2']

        
    X_1324 = X_1024[['tem_loc3','hum_loc3','coil_loc3','tem_out','hum_out','month','day','hour']]
    X_1324.columns=['tem_in','hum_in','tem_coil','tem_out_loc1','hum_out_loc1','month','day','hour']
    y_1324 = y_1024['X24H_cond_loc3']
        
    X_1348 = X_1048[['tem_loc3','hum_loc3','coil_loc3','tem_out','hum_out','month','day','hour']]
    X_1348.columns=['tem_in','hum_in','tem_coil','tem_out_loc1','hum_out_loc1','month','day','hour']
    y_1348 = y_1048['X48H_cond_loc3']

        
    X_2124 = X_2024[['tem_loc1','hum_loc1','coil_loc1','tem_out','hum_out','month','day','hour']]
    X_2124.columns=['tem_in','hum_in','tem_coil','tem_out_loc1','hum_out_loc1','month','day','hour']
    y_2124 = y_2024['X24H_cond_loc1']

    
    X_2148 = X_2048[['tem_loc1','hum_loc1','coil_loc1','tem_out','hum_out','month','day','hour']]
    X_2148.columns=['tem_in','hum_in','tem_coil','tem_out_loc1','hum_out_loc1','month','day','hour']
    y_2148 = y_2048['X48H_cond_loc1']


    X_2224 = X_2024[['tem_loc2','hum_loc2','coil_loc2','tem_out','hum_out','month','day','hour']]
    X_2224.columns=['tem_in','hum_in','tem_coil','tem_out_loc1','hum_out_loc1','month','day','hour']
    y_2224 = y_2024['X24H_cond_loc2']


    X_2248 = X_2048[['tem_loc2','hum_loc2','coil_loc2','tem_out','hum_out','month','day','hour']]
    X_2248.columns=['tem_in','hum_in','tem_coil','tem_out_loc1','hum_out_loc1','month','day','hour']
    y_2248 = y_2048['X48H_cond_loc2']

    
    X_2324 = X_2024[['tem_loc3','hum_loc3','coil_loc3','tem_out','hum_out','month','day','hour']]
    X_2324.columns=['tem_in','hum_in','tem_coil','tem_out_loc1','hum_out_loc1','month','day','hour']
    y_2324 = y_2024['X24H_cond_loc3']

    X_2348 = X_2048[['tem_loc3','hum_loc3','coil_loc3','tem_out','hum_out','month','day','hour']]
    X_2348.columns=['tem_in','hum_in','tem_coil','tem_out_loc1','hum_out_loc1','month','day','hour']
    y_2348 = y_2048['X48H_cond_loc3']
  
    
 
    tree1124.fit(X_1124, y_1124.values.ravel())
   
    tree1148.fit(X_1148, y_1148.values.ravel())
   
    tree1224.fit(X_1224, y_1224.values.ravel())
   
    tree1248.fit(X_1248, y_1248.values.ravel())
   
    tree1324.fit(X_1324, y_1324.values.ravel())
   
    tree1348.fit(X_1348, y_1348.values.ravel())
 
    tree2124.fit(X_2124, y_2124.values.ravel())
 
    tree2148.fit(X_2148, y_2148.values.ravel())
  
    tree2224.fit(X_2224, y_2224.values.ravel())
 
    tree2248.fit(X_2248, y_2248.values.ravel())
 
    tree2324.fit(X_2324, y_2324.values.ravel())
    
    tree2348.fit(X_2348, y_2348.values.ravel())
    
```

```python
def predict_data(url):
    data = pd.read_csv(url,parse_dates=['plant_test.mea_ddhr'])
    
    data["month"] = data["plant_test.mea_ddhr"].dt.month
    data["day"] = data["plant_test.mea_ddhr"].dt.day
    data["hour"] = data["plant_test.mea_ddhr"].dt.hour
    
    vali_data = pd.read_csv('./validation.csv')
    print(vali_data.head(5))
    for i in data.index:
        plant = data.loc[i]['plant_test.plant']
        loc = data.loc[i]['plant_test.loc']
        model_24 = int(plant)*1000 + int(loc)*100 + 24
        model_48 = int(plant)*1000 + int(loc)*100 + 48
        
        feature = ['plant_test.tem_in','plant_test.hum_in','plant_test.tem_coil',
                   'plant_test.tem_out_loc1','plant_test.hum_out_loc1','month','day','hour']
        
        X_test = data.loc[i, feature]
        X_test.columns = ['tem_in','hum_in','tem_coil','tem_out_loc1','hum_out_loc1','month','day','hour']
        
        
        if model_24 == 1124:
            pred24 = tree1124.predict([X_test])
            pred24_proba = tree1124.predict_proba([X_test])
            
        elif model_24 == 2124:
            pred24 = tree2124.predict([X_test])
            pred24_proba = tree2124.predict_proba([X_test])
                
        elif model_24 == 1224:
            pred24 = tree1224.predict([X_test])
            pred24_proba = tree1224.predict_proba([X_test])
            
        elif model_24 == 2224:
            pred24 = tree2224.predict([X_test])
            pred24_proba = tree2224.predict_proba([X_test])
            
        elif model_24 == 1324:
            pred24 = tree1324.predict([X_test])
            pred24_proba = tree1324.predict_proba([X_test])
            
        elif model_24 == 2324:
            pred24 = tree2324.predict([X_test])
            pred24_proba = tree2324.predict_proba([X_test])
            
            
            
        if model_48 == 1148:
            pred48 = tree1148.predict([X_test])
            pred48_proba = tree1148.predict_proba([X_test])
            
        elif model_48 == 2148:
            pred48 = tree2148.predict([X_test])
            pred48_proba = tree2148.predict_proba([X_test])
            
        elif model_48 == 1248:
            pred48 = tree1248.predict([X_test])
            pred48_proba = tree1248.predict_proba([X_test])
            
        elif model_48 == 2248:
            pred48 = tree2248.predict([X_test])
            pred48_proba = tree2248.predict_proba([X_test])
            
        elif model_48 == 1348:
            pred48 = tree1348.predict([X_test])
            pred48_proba = tree1348.predict_proba([X_test])
            
        elif model_48 == 2348:
            pred48 = tree2348.predict([X_test])
            pred48_proba = tree2348.predict_proba([X_test])
        
#         y_test_24 = data['plant_test.x24h_cond_loc']
#         y_test_24.columns = ['x24h_cond']
#         y_test_48 = data['plant_test.x48h_cond_loc']
#         print(data.loc[i]['plant_test.x24h_tma'])
        print(vali_data.loc[i])
    
        if pred24_proba[0][0] > pred24_proba[0][1]:
            vali_data['X24H_COND_LOC'][i] = 0
        else:
            vali_data['X24H_COND_LOC'][i] = 1
            
        if pred48_proba[0][0] > pred48_proba[0][1]:
            vali_data['X48H_COND_LOC'][i] = 0         
        else:
            vali_data['X48H_COND_LOC'][i] = 1

        vali_data['X24H_COND_LOC_PROB'][i] = pred24_proba[0][int(vali_data['X24H_COND_LOC'][i])]*100
        vali_data['X48H_COND_LOC_PROB'][i] = pred48_proba[0][int(vali_data['X48H_COND_LOC'][i])]*100
        
#         print("plant : "+str(plant)+" loc : "+str(loc))
#         print(pred24, pred48)
#         print(pred24_proba, pred48_proba)
#         print()
    vali_data.to_csv('./validation_predict_normalize_svm_day_c100_gamma10.csv')
```
