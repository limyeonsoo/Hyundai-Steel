```python
import pandas as pd

plant1 = pd.read_csv('./plant1_train.csv',parse_dates=['plant1_train.mea_ddhr'])
plant1.head(10)
```

```python
del plant1['Unnamed: 0']
```

```python
plant1["year"] = plant1["plant1_train.mea_ddhr"].dt.year
plant1["month"] = plant1["plant1_train.mea_ddhr"].dt.month
plant1["day"] = plant1["plant1_train.mea_ddhr"].dt.day
plant1["hour"] = plant1["plant1_train.mea_ddhr"].dt.hour
plant1["minute"] = plant1["plant1_train.mea_ddhr"].dt.minute
plant1["second"] = plant1["plant1_train.mea_ddhr"].dt.second
plant1.head()
```

```python
plant1_del = plant1.dropna(axis=0)
```

```python
Y1 = plant1_del[['plant1_train.cond_loc1']]
Y2 = plant1_del[['plant1_train.cond_loc2']]
Y3 = plant1_del[['plant1_train.cond_loc3']]
Y = plant1_del[['plant1_train.cond_loc1','plant1_train.cond_loc2','plant1_train.cond_loc3']]
X = plant1_del.drop(['plant1_train.mea_ddhr','plant1_train.cond_loc1','plant1_train.cond_loc2','plant1_train.cond_loc3'],axis=1)
```

```python
from sklearn.model_selection import train_test_split

X1_train, X1_test, Y1_train, Y1_test = train_test_split(X,Y1,test_size=0.2,shuffle=True)
X2_train, X2_test, Y2_train, Y2_test = train_test_split(X,Y2,test_size=0.2,shuffle=True)
X3_train, X3_test, Y3_train, Y3_test = train_test_split(X,Y3,test_size=0.2,shuffle=True)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.naive_bayes import GaussianNB
```

```python
X1_train = X1_train.values
X2_train = X2_train.values
X3_train = X3_train.values
Y1_train = Y1_train.values.ravel()
Y2_train = Y2_train.values.ravel()
Y3_train = Y3_train.values.ravel()
```

```python
naive1 = GaussianNB()
```

```python
naive1.fit(X1_train,Y1_train)
```

```python
from sklearn import metrics
```

```python
pred1 = naive1.predict(X1_test)
print("loc1 훈련 세트 정확도: {:.3f}".format(naive1.score(X1_train, Y1_train)))
print("loc1 테스트 세트 정확도: {:.3f}".format(metrics.roc_auc_score(Y1_test,pred1)))
```

```python
naive2 = GaussianNB()
naive2.fit(X2_train,Y2_train)
pred2 = naive2.predict(X2_test)
print("loc2 훈련 세트 정확도: {:.3f}".format(naive2.score(X2_train, Y2_train)))
print("loc2 테스트 세트 정확도: {:.3f}".format(metrics.roc_auc_score(Y2_test,pred2)))
```

```python
naive3 = GaussianNB()
naive3.fit(X3_train,Y3_train)
pred3 = naive3.predict(X3_test)
print("loc3 훈련 세트 정확도: {:.3f}".format(naive3.score(X3_train, Y3_train)))
print("loc3 테스트 세트 정확도: {:.3f}".format(metrics.roc_auc_score(Y3_test,pred3)))
```
